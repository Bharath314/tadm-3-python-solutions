### Exercise 1-5

**Q.** The knapsack problem is as follows: given a set of integers $S = \{s_1 , s_2 , . . . , s_n \}$,
and a target number $T$ , ﬁnd a subset of $S$ that adds up exactly to $T$. For example,
there exists a subset within $S = \{1, 2, 5, 9, 10\}$ that adds up to $T = 22$
but not $T = 23$.

Find counterexamples to each of the following algorithms for the knapsack problem.
That is, give an $S$ and $T$ where the algorithm does not ﬁnd a solution that
leaves the knapsack completely full, even though a full-knapsack solution exists.

**(a).** Put the elements of $S$ in the knapsack in left to right order if they ﬁt, that
is, the ﬁrst-ﬁt algorithm.

The first-fit algorithm can be implemented as follows:

````python
def first_fit(S: list[int], T: int):
    knapsack = []
    current_sum = 0
    for number in S:
        if number <= (T - current_sum):
            knapsack.append(number)
            current_sum += number
        if current_sum == T:
            break
    if current_sum == T:
        return knapsack
    return []
````

Take $S = \{1, 2, 3, 4\}$ and $T = 7$. The full-knapsack solution would be $\{1, 2, 4\}$.  However, upon running the first-fit algorithm on these values, we fail to find the solution.

````python
first_fit([1, 2, 3, 4], 7)
````
