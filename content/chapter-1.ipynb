### Exercise 1-5

**Q.** The knapsack problem is as follows: given a set of integers $S = \{s_1 , s_2 , . . . , s_n \}$,
and a target number $T$ , ﬁnd a subset of $S$ that adds up exactly to $T$. For example,
there exists a subset within $S = \{1, 2, 5, 9, 10\}$ that adds up to $T = 22$
but not $T = 23$.

Find counterexamples to each of the following algorithms for the knapsack problem.
That is, give an $S$ and $T$ where the algorithm does not ﬁnd a solution that
leaves the knapsack completely full, even though a full-knapsack solution exists.

**(a).** Put the elements of $S$ in the knapsack in left to right order if they ﬁt, that
is, the ﬁrst-ﬁt algorithm.

The first-fit algorithm can be implemented as follows:

````python
def first_fit(S: list[int], T: int):
    knapsack = []
    current_sum = 0
    for number in S:
        if number <= (T - current_sum):
            knapsack.append(number)
            current_sum += number
        if current_sum == T:
            break
    if current_sum == T:
        return knapsack
    return []
````

Take $S = \{1, 2, 3, 4\}$ and $T = 7$. The full-knapsack solution would be $\{1, 2, 4\}$.  However, upon running the first-fit algorithm on these values, we fail to find the solution.

````python
first_fit([1, 2, 3, 4], 7)
````

**(b).** Put the elements of S in the knapsack from smallest to largest,
that is, the best-ﬁt algorithm.

The best-fit algorithm can be implemented as follows:

````python
def best_fit(S: list[int], T: int):
    items = list(S)
    knapsack = []
    current_sum = 0
    while current_sum < T:
        gap = T - current_sum
        best_candidate = None
        for x in items:
            if x <= gap:
                if best_candidate is None or x > best_candidate:
                    best_candidate = x
        if best_candidate is not None:
            knapsack.append(best_candidate)
            current_sum += best_candidate
            items.remove(best_candidate)
        else:
            break
   
    if current_sum == T:
        return knapsack
    return []
````

Take $S = \{2, 3, 3, 4, 5\}$ and $T = 10$. The full-knapsack solution would be $\{2, 3, 5\}$.  However, upon running the best-fit algorithm on these values, we fail to find the solution.

````python
best_fit([2, 3, 3, 4, 5], 10)
````

**(c).** Put the elements of S in the knapsack from largest to smallest.

This algorithm can be implemented as follows:

````python
def first_fit_sorted(S: list[int], T: int):
    S = sorted(S, reverse=True)
    knapsack = []
    current_sum = 0
    for number in S:
        if number <= (T - current_sum):
            knapsack.append(number)
            current_sum += number
        if current_sum == T:
            break
    if current_sum == T:
        return knapsack
    return []
````

Take $S = \{2, 3, 3, 4, 5\}$ and $T = 10$. The full-knapsack solution would be $\{2, 3, 5\}$.  However, upon running the sorted first-fit algorithm on these values, we fail to find the solution.

````python
first_fit_sorted([2, 3, 3, 4, 5], 10)
````

## Exercise 1-6

**Q.** The set cover problem is as follows: given a set $S$ of subsets $S_1 , \dots, S_m$ of
the universal set $U = \{1, \dots, n\}$, find the smallest subset of subsets $T ⊆ S$ such
that $\cup_{t_i ∈ T}t_i = U$ . For example, consider the subsets $S_1 = \{1, 3, 5\}$, $S_2 = \{2, 4\}$, $S_3 = \{1, 4\}$, and $S_4 = \{2, 5\}$. The set cover of $\{1, \dots, 5\}$ would then be $S_1$ and $S_2$.

Find a counterexample for the following algorithm: Select the largest subset for
the cover, and then delete all its elements from the universal set. Repeat by
adding the subset containing the largest number of uncovered elements until all
are covered.

The algorithm can be implemented as follows:

````python
def greedy_set_cover(S: list[set[int]]):
    U = set().union(*S)
    cover = []
    while U:
        best_set = max(S, key=lambda s: len(s & U))
        cover.append(best_set)
        U -= best_set
        S.remove(best_set)
    return cover
````

Take $S = \{\{1,3,5\}, \{1,2\}, \{3,4\}, \{5,6\}\}$ and $T = 10$. The set cover would be $\{\{1,2\}, \{3,4\}, \{5,6\}\}$.  However, upon running the greedy set cover algorithm on this set, we fail to find the solution.

````python
greedy_set_cover([{1,3,5}, {1,2}, {3,4}, {5,6}])
````

**Q.** The maximum clique problem in a graph $G = (V, E)$ asks for the largest
subset $C$ of vertices $V$ such that there is an edge in $E$ between every pair of
vertices in $C$. Find a counterexample for the following algorithm: Sort the
vertices of $G$ from highest to lowest degree. Considering the vertices in order of degree, for each vertex add it to the clique if it is a neighbor of all vertices
currently in the clique. Repeat until all vertices have been considered.

The algorithm can be implemented as follows:

````python
def maximum_clique(G: tuple[set[str], list[tuple[str, str]]]):
    vertices = G[0]
    edges = G[1]
    graph_degrees = {vertex: 0 for vertex in vertices}
    neighbours: dict[str, set] = {vertex: set() for vertex in vertices}
    for pair in edges:
        graph_degrees[pair[0]] += 1
        graph_degrees[pair[1]] += 1
        neighbours[pair[0]].add(pair[1])
        neighbours[pair[1]].add(pair[0])

    vertices_ordered_by_degree = sorted(
        list(vertices), key=lambda vertex: graph_degrees[vertex], reverse=True
    )
    print(vertices_ordered_by_degree)
    print(neighbours)
    clique = set()
    for vertex in vertices_ordered_by_degree:
        clique_member = True
        for member in clique:
            if vertex not in neighbours[member]:
                clique_member = False
                break
        if clique_member:
            clique.add(vertex)
    return clique
````

Take graph $G=(V,E)$, where $V=\{A,W,X,Y,Z,D,E,F\}$ and $E=\{(A,W),(A,X),(A,Y),(A,Z),(D,E),(E,F),(F,D)\}$

```mermaid

    A --- Z

    E --- F
    F --- D
```

The maximum clique would be $\{D,E,F\}$. However, upon running the algorithm on this graph, we fail to find the solution

### Exercise 1-8

**Q.** Prove the correctness of the following recursive algorithm to multiply two
natural numbers, for all integer constants $c ≥ 2$.

$$
\begin{aligned}
&\text{Multiply}(y, z) \\
&\quad \textit{if } z = 0 \textit{ then } \text{return}(0) \textit{ else} \\
&\quad \text{return}(\text{Multiply}(cy, \lfloor z/c \rfloor) + y \cdot (z \bmod c))
\end{aligned}
$$

**Base Case:**

Taking $z=0$, $\text{Multiply}(y,z) = 0$, which is correct as $y \times 0 = 0$

Thus, $\text{Multiply}$ is proven to work on the basis case


**Inductive step:**

To prove the correctness of the function for $y = y^*$ and $z = z^*$, we assume that the function works correctly for the next recursive call where $y = cy^*$ and $z = \lfloor z^*/c \rfloor$, *i.e.*, $\text{Multiply}(cy^*, \lfloor z^*/c \rfloor) = cy^* \times \lfloor z^*/c \rfloor$.

$$
\begin{align*}
\text{Multiply}(y^*, z^*) &= \text{Multiply}(cy^*, \lfloor z^*/c \rfloor) + y^* \cdot (z^* \bmod c) \\
&= cy^* \cdot \lfloor z^*/c \rfloor + y^* \cdot (z^* \bmod c) \\
&= cy^* \cdot \frac{z^* - z^* \bmod c}{c} + y^* \cdot (z^* \bmod c) \\
&= y^* z^* - y^*(z^* \bmod c) + y^* \cdot (z^* \bmod c) \\
&= y^* z^*
\end{align*}
$$

Thus, by induction, we have proved the correctness of $\text{Multiply}(y,z)$
