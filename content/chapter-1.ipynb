## Finding Counterexamples

### Exercise 1-1

**Q.** Show that $a + b$ can be less than $min(a, b)$

If we take $a$ to be $-3$ and $b$ to be $-4$, $a + b$ equals $-7$, which is less than either number.

### Exercise 1-2

**Q.** Show that $a \times b$ can be less than $min(a, b)$.

If we take $a$ to be $2$ and $b$ to be $-2$, $a \times b$ equals $-4$, which is less than either number.
